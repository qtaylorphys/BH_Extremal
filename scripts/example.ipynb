```python
# Import necessary modules
from pbh import PrimordialBlackHole
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define parameters
spacetime = "Kerr"
initial_mass = 100.
final_mass = 1.
initial_momentum = 0.
save_path = True
```

```python
# Create PrimordialBlackHole object
pbh = PrimordialBlackHole(
    spacetime,
    initial_mass, final_mass,
    initial_momentum,
    save_path=save_path,
)

# Evolve the black hole
pbh.evolve()
```

```python
# Generate steps for plotting
steps = np.arange(pbh.path.shape[0])

# Plot mass evolution
plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 0])

plt.xlabel(r"step $n$")
plt.ylabel(r"mass M(n)")

plt.show()

# Plot angular momentum evolution
plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 1])

plt.xlabel(r"step $n$")
plt.ylabel(r"angular momentum J(n)")

plt.show()

# Plot a_* (spin parameter) evolution
plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 2])

plt.xlabel(r"step $n$")
plt.ylabel(r"a_{*} (n)")

plt.show()

# Plot temperature evolution
plt.figure(figsize=(8, 6))
plt.plot(steps, pbh.path[:, 4])

plt.yscale('log')

plt.xlabel(r"step $n$")
plt.ylabel(r"temperature T(n)")

plt.show()
```
